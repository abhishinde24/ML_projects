```python
import opendatasets as od

download_url = 'https://raw.githubusercontent.com/jbrownlee/Datasets/master/pima-indians-diabetes.data.csv'

od.download(download_url)
```

```python
data_filename = '/content/pima-indians-diabetes.data.csv'
```

```python
import pandas as pd
import numpy as np
```

```python
df = np.loadtxt(data_filename,delimiter=',')
```

```python
x = df[:,:8]
y = df[:,8]
```

## defining keras model

```python
from keras.models import Sequential
from keras.layers import Dense
```

```python
model = Sequential()
model.add(Dense(12, input_dim=8, activation='relu'))
model.add(Dense(8, activation='relu'))
model.add(Dense(1, activation='sigmoid'))
```

```python

# compile the keras model
model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# fit the keras model on the dataset
model.fit(x, y, epochs=150, batch_size=10)
```

```python
# evaluate the keras model
# here we are evaluating model on train data only we have not created train-test split
_, accuracy = model.evaluate(x, y)
print('Accuracy: %.2f' % (accuracy*100))
```

- model is giving accuracy of about 77.7 %
